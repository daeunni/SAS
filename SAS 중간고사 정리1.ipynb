```sas
*dlm : 구분자, infile 문에 넣기!;
*짤리니까 length 지정해주기 ^____^ 근데 변수 순서 바뀜;

data subset3;
    infile 'sales.txt' dlm=',';
    length job_title$13;
    input Employid first_name$ last_name$ gender$ salary job_title$ country$;
run;
proc print data=subset3;
run;
```

```sas
* 변수 순서 안바뀌게 하기;
data subset;
    infile 'sales.txt' dlm=',';
    length first_name$10 job_title$13;
    input Employid first_name$ last_name$ gender$ salary job_title$ country$;
run;

proc print data=subset;
run;
```

```sas
*retain으로 변수 순서 다시 지정!! ;;;;;;;;

data a;
    retain Employid first_name last_name gender salary job_title country;
    set subset;
    *subset 복사
run;

proc print data=a;
run;
```

```sas
data subset3;
    infile 'sales1.txt' dlm=',';
    input Employid first_name$ last_name$ gender$ salary job_title$ country$ 
    birthday :date. hire_date :mmddyy10.
    
;run;
proc print data=subset3;
run;
```

```sas
* 구분자 & 인경우 + 데이터에 변수 포함된경우 ;
* firstobs 이용해도 되는듯. dlm은 &로 지정!;

data strange;
    infile 'special.txt' dlm='&' firstobs=2;
    input Region$ State$ Month$ Expenses Revenue;
run;
proc print data=strange;
run;
```

```sas
*proc 단계에서 시행하기;
proc import datafile='special.txt' out=mydata dbms=dlm replace;
delimiter='&';
getnames=yes;
run;

options nodate ps=60 ls=80;
proc print data=mydata;
run;
```

#### < 문자 함수 >

```sas
*trim : 뒤 공백 제거 + || : 문자열 결합함수;
data a1;
    input a $ b $ @@;
    cards;
a b 
a b 
a b 
;
run;

proc print data=a1;
run;
```

```sas
data a2;
    set a1;
    c = a || b;
    *그냥 문자형 변수 결합함;
run;

proc print data=a2;
run;
```

```sas
*trim 넣어서 결합하면 공백이 사라짐!;;;;
data a3;
    set a1;
    c = trim(a) || b;
run;

proc print data=a3;
run;
```

### < 형 변환 함수 >

```sas
data test;
    input sale $9. ;
    fmtsale = input(sale, comma9.);
    *input 활용해 숫자변수로 변환!;
cards;
2,115,353
;
run;
proc print;
run;
```

```sas
*put : 숫자 > 문자 이용
data test2;
    input ymd 8.;
    fmtymd = put(ymd, $8.);
    cyear = substr(fmtymd, 1, 4);
    nyear = input(cyear, 4.);
cards;
20200305
;run;

proc print;
run;
```

```sas
*주민번호 성별 추출 : substr 이용;
data number;
    id = '001010-1234567';
    gender = substr(id, 8, 1);
put id '---->' gender;
proc print data=number;

*put : 계산 로그 출력(sas 프로그램에서 하면 됨)
```

```sas
* 주민번호로 현재나이 계산;
data number2;
    id = '031212-1234567';
    birth_year = 2000 + substr(id, 1, 2);
    today = date();
    *현재 날짜 나타낼 땐 date() 이용 ! 그거의 년도는 year()에 넣기!!
    today_year = year(today);
    current_age = today_year -birth_year + 1;
    
put birth_year today_year current_age;
run;

proc print;
run;
```

```sas
data data;
    input name &$ 18. indate 5. salary 8.;
    format indate yymmdd8. salary comma8.;
    cards;
Martin, Virginia  7526 34800
Singleton, MaryAnn  9245 27900
leighton, Morice  8750 32600
Freeler, Carl  10272 29900
Cage, Merce  8327 39800
;
run;

proc print data=data;
run;
```

```sas
*length로 길이 먼저 지정해주기! ;
data data;
    length name $18 indate 8 salary 8;
    input name &$ indate salary;
    format indate yymmdd8. salary comma8.;
    cards;
Martin, Virginia  7526 34800
Singleton, MaryAnn  9245 27900
leighton, Morice  8750 32600
Freeler, Carl  10272 29900
Cage, Merce  8327 39800
;
run;

proc print data=data;
run;
```

```sas
data ex_2;
    input name $10. ;
    cards;
강 미경
전 예지
김 현솔
;

proc print;
run;
```

```sas
* substr 바이트로 적용됨을 주의하자!!!!!!! ;;;;;;;;;;;;;;;;;;;;;
* keep 사용! 
* catx : 앞 뒤 공백 제거하고 구분자로 합치기(구분자, 앞, 뒤 순서! )
data ex_3(keep = name change);
    set ex_2;
    sung = substr(name, 1, 3);
    new_name = substr(name, 5);
    change = catx(' ', new_name, sung);
run;

proc print data=ex_3;
run;
```

```sas
* substr byte 단위임을 완전 주의해야겠다!!!!!!!!!!!;;;;;;;;;;;;;;

data ex_3;
    set ex_2;
    sung = substr(name, 1, 3);
    new_name = substr(name, 5, 6);
    change = catx(' ', new_name, sung);
run;

proc print data=ex_3;
run;
```
